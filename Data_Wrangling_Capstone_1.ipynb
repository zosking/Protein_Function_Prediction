```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

### 1. Generate a list of Swiss-Prot reviewed human protein accession numbers.

```python
# make list of verified accession numbers of human proteins
url = 'https://www.ebi.ac.uk/proteins/api/proteomes/proteins/UP000005640?reviewed=true'

human_proteome = requests.get(url, headers={ "Accept" : "application/json"})
```

```python
human = human_proteome.json()
accession_df = json_normalize(data = human['component'], record_path = ['protein'])
accession_numbers = list(accession_df.accession)
```

### 3. Create a list of dictionaries containing all of the protein data.

```python
import glob

path = '/Users/zosking/Desktop/SpringBoard_Projects/Capstone_1/Data_Wrangling/*.txt'
files = glob.glob(path)
all_dict = {}

for name in files:
    with open(name, 'r') as filehandle:
        flat_list = json.load(filehandle)
        for x in range(len(flat_list)):
            all_dict.update({flat_list[x][0]['accession']:flat_list[x][0]})           
```

```python
human_proteins = pd.DataFrame.from_dict(all_dict, orient='index')
human_proteins
```

### 4. Unnesting protein data.

```python
from pandas.io.json import json_normalize 

sequence_info = json_normalize(data= human_proteins['sequence'])
human = human_proteins.reset_index()
human.index
```

```python
merged = human.merge(sequence_info, left_index=True, right_index=True)
merged.head()
```

```python
merged.to_csv('full_proteins.csv')
```

```python
%%time

features_df=pd.DataFrame()
count = 0
feats = merged['features']

for x in range(len(merged)):
    df = json_normalize(feats[x])
    features_df = features_df.append(df, ignore_index=True)
    count +=1
    print(str(count) + ' '+ 'df' + ' appended')
```

```python
ls = []
z=0
for x in range(len(merged)):
    df = json_normalize(merged['features'][x])
    z += 1
    for y in range(len(df)):
        ls.append(z)
```

```python
features_df['number'] = ls
```

```python
key = {}
z = 0
for x in range(len(merged.accession)):
    z += 1
    key.update({z:merged.accession[x]})
```

```python
features_df['new_in'] = features_df['number'].map(key)
features = features_df.set_index('new_in')
features.to_csv('features.csv')
```

```python
category_count = features.groupby(['new_in','category'])['number'].count().unstack().fillna(0)
type_count = features.groupby(['new_in','type'])['number'].count().unstack().fillna(0)
```

```python
feature_counts = category_count.merge(type_count, left_index=True, right_index=True)
```

```python
com = merged[merged['comments'].isna() == False]
com = com.reset_index()
```

```python
%%time

comments_df=pd.DataFrame()
count = 0

for x in range(len(merged)):
    df = json_normalize(com['comments'][x])
    comments_df = comments_df.append(df, ignore_index=True)
    count +=1
    print(str(count) + ' '+ 'df' + ' appended')
    
comments_df.to_csv('comments_df.csv')
```

```python
%%time
com = com.set_index('accession')
locations_df=pd.DataFrame()
count = 0

for x in range(len(com['comments'])):
    ind = com['comments'].index[x]
    for y in range(len(com['comments'][x])):
        try:        
            if com['comments'][x][y]['type'] == 'SUBCELLULAR_LOCATION':
                loc_list = com['comments'][x][y]['locations']
                for z in range(len(loc_list)):
                    location = loc_list[z]['location']['value']
                    df = pd.DataFrame(data=[(ind, location,1)], columns = ['accession', 'location','number'])
                    locations_df = locations_df.append(df, ignore_index=True)
                    count +=1
                    print(str(count) + ' '+ 'df' + ' appended')
            else:
                pass
        except:
            pass    
        
locations_df.to_csv('locations_df.csv')
```

```python
%%time

ls = []
z=0
for x in range(len(com)):
    df = json_normalize(com['comments'][x])
    z += 1
    for y in range(len(df)):
        ls.append(z)
```

```python
comments_df['number'] = ls
```

```python
m1 = merged.set_index('accession')
merged_features = m1.merge(feature_counts, left_index=True, right_index=True)
```

```python
merged_features2 = merged_features.merge(locations_counts, left_index=True, right_index=True)
```

```python
%%time

keywords_df=pd.DataFrame()
count = 0

for x in range(len(merged_features2)):
        df = json_normalize(merged_features2['keywords'][x])
        keywords_df = keywords_df.append(df, ignore_index=True)
        count +=1
        print(str(count) + ' '+ 'df' + ' appended')
    
keywords_df.to_csv('keywords.csv')
```

```python
%%time

ls = []
z=0
for x in range(len(merged_features2)):
    df = json_normalize(merged_features2['keywords'][x])
    z += 1
    for y in range(len(df)):
        ls.append(z)
```

```python
keywords_df['number'] = ls
```

```python
key = {}
z = 0
for x in range(len(merged_features2['index'])):
    z += 1
    key.update({z:merged_features2['index'][x]})
    

keywords_df['new_in'] = keywords_df['number'].map(key)
keywords = keywords_df.set_index('new_in')
keywords
```

```python
keyword_counts = keywords.groupby(['new_in','value'])['number'].count().unstack().fillna(0)
```

```python
proteins_df = merged_features2.merge(keyword_counts, left_index=True, right_index=True)
```

```python
drops = ['index','geneLocations','organism','id','proteinExistence','info','secondaryAccession','protein','gene','comments','features','dbReferences','keywords','references','sequence_x']
```

```python
proteins_clean = proteins_df.drop(drops, axis=1)
```

```python
proteins_clean['date'] = pd.to_datetime(proteins_clean['modified'])
```

```python
from collections import Counter

seq_data = pd.DataFrame()

for x in range(len(proteins_clean['sequence_y'])):
    i = Counter(proteins_clean['sequence_y'][x])
    seq_df = pd.DataFrame.from_dict(i,orient='index')
    seq_data = seq_data.append(i, ignore_index=True)
```

```python
seq_data.index
```

```python
proteins_clean = proteins_clean.reset_index()
df = proteins_clean.merge(seq_data, left_index=True, right_index=True)
```

```python
positive = ['R','H','K']
negative = ['D','E']
nonpolar = ['G','A','V','C','P','L','I','M','W','F']
polar = ['S','T','Y','N','Q']

df['positive_aa'] = df[positive].sum(axis=1)
df['negative_aa'] = df[negative].sum(axis=1)
df['nonpolar_aa'] = df[nonpolar].sum(axis=1)
df['polar_aa'] = df[polar].sum(axis=1)
```

```python
import re

mito_proteins = df[df.loc(axis=1)[df.columns.str.contains('Mitochondr', re.IGNORECASE)].values == 1.0]
nuclear_proteins = df[df.loc(axis=1)[df.columns.str.contains('Nucleus', re.IGNORECASE)].values == 1.0]
membrane_proteins = df[df.loc(axis=1)[df.columns.str.contains('Cell membrane', re.IGNORECASE)].values == 1.0]
cytoplasm_proteins = df[df.loc(axis=1)[df.columns.str.contains('Cytoplasm,', re.IGNORECASE)].values == 1.0]
cytoskeletal_proteins = df[df.loc(axis=1)[df.columns.str.contains('Cytoplasm, cytoskeleton', re.IGNORECASE)].values == 1.0]
vesicle_proteins = df[df.loc(axis=1)[df.columns.str.contains('Cytoplasmic vesicle', re.IGNORECASE)].values == 1.0]
er_proteins = df[df.loc(axis=1)[df.columns.str.contains('Endoplasmic reticulum', re.IGNORECASE)].values == 1.0]
endosome_proteins = df[df.loc(axis=1)[df.columns.str.contains('Endosome', re.IGNORECASE)].values == 1.0]
golgi_proteins = df[df.loc(axis=1)[df.columns.str.contains('Golgi', re.IGNORECASE)].values == 1.0]
```

```python
mito_proteins['loc'] = 'mitochondria'
nuclear_proteins['loc'] = 'nucleus'
membrane_proteins['loc'] = 'membrane'
cytoplasm_proteins['loc'] = 'cytoplasm'
cytoskeletal_proteins['loc'] = 'cytoskeleton'
vesicle_proteins['loc'] = 'vesicle'
er_proteins['loc'] = 'endoplasmic_reticulum'
endosome_proteins['loc'] = 'endosome'
golgi_proteins['loc'] = 'golgi'
```

```python
mito_proteins = mito_proteins.drop_duplicates(keep='first')
nuclear_proteins = nuclear_proteins.drop_duplicates(keep='first')
membrane_proteins = membrane_proteins.drop_duplicates(keep='first')
cytoplasm_proteins = cytoplasm_proteins.drop_duplicates(keep='first')
cytoskeletal_proteins = cytoskeletal_proteins.drop_duplicates(keep='first')
vesicle_proteins = vesicle_proteins.drop_duplicates(keep='first')
er_proteins = er_proteins.drop_duplicates(keep='first')
endosome_proteins = endosome_proteins.drop_duplicates(keep='first')
golgi_proteins = golgi_proteins.drop_duplicates(keep='first')
```

```python
locs = [mito_proteins,nuclear_proteins,membrane_proteins,cytoplasm_proteins,cytoskeletal_proteins,vesicle_proteins,er_proteins,endosome_proteins,golgi_proteins]
```

```python
proteins_df_final = pd.concat(locs, join='outer')
```

```python
proteins_df_final.to_csv('proteins_df_final.csv')
```
